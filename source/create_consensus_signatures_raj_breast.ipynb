```r
library(tidyverse)
```

```r

raj_vs_control_de_signature <- readRDS("/data/CDSL_hannenhalli/Cole/projects/drug_treatment/data/genesets/raj_breast_clusters_vs_control_de_genesets_500.rds")

raj_de_results <- readRDS("/data/CDSL_hannenhalli/Cole/projects/drug_treatment/data/de_results/raj_breast_cluster_de.rds")
```

```r
all_clusters <- as.numeric(unique(raj_de_results$cluster))


temp <- list()
for(i in all_clusters){
    curr_cluster_de_genes <- raj_de_results %>% 
    filter(p_val_adj < 0.05 & avg_log2FC > 0 & cluster == i)  %>% 
    pull(gene)
    
    curr_signature <- intersect(raj_vs_control_de_signature[[i]], curr_cluster_de_genes)
    
    temp <- append(temp,list(curr_signature))
    
}
```

```r
saveRDS(temp, "/data/CDSL_hannenhalli/Cole/projects/drug_treatment/data/genesets/raj_breast_cluster_and_control_consensus_signatures.rds")
```
